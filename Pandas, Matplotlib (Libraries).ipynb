# Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Histogram

```python
plt.hist(x=df['Age'],color="hotpink")
plt.title("Distribution of age",fontsize=20,color='blue')
plt.xlabel("Age",fontsize=15,color='green')
plt.ylabel("Frequency",fontsize=15,color='green')
plt.show()
```

### Boxplot

```python
#seaborn
sns.boxplot(x=df['Age'],color="orange")
plt.title("Age",fontsize=20,color='blue')
plt.xlabel("Age",fontsize=15,color='green')
plt.show()
```

### Barplot

```python
df.Sex.value_counts().plot(kind="bar")
plt.title("sex disribution",fontsize=20,color="blue")
plt.xlabel("sex",fontsize=15,color="green")
plt.ylabel("count",fontsize=15,color="green")
plt.show()
```

```python
df.Sex.value_counts().plot(kind="barh",color="green")
plt.title("sex disribution",fontsize=20,color="blue")
plt.xlabel("sex",fontsize=15,color="green")
plt.ylabel("count",fontsize=15,color="green")
plt.show()
```

```python
df.Pclass.value_counts().plot(kind="bar",color="red")
plt.title("Passenger",fontsize=20,color="blue")
plt.xlabel("Passenger class",fontsize=15,color="green")
plt.show()
```

```python
df.Pclass.value_counts().plot(kind="barh",color="brown")
plt.title("Passenegr",fontsize=20,color="blue")
plt.xlabel("Passenger class",fontsize=15,color="green")

plt.show()
```

### Pie chart

```python
e=df['Embarked'].value_counts()
keys=e.keys().to_list()
c=e.to_list()
plt.pie(x=c,labels=keys,autopct='%1.1f%%')
plt.show()
```
